```python
class TiendaOnline:
    def __init__(self):
        self.inventario = []  # inventario = [{'nombre': 'Camisa', 'precio': 20, 'cantidad': 40}, {'nombre': 'Pantal칩n', 'precio': 30, 'cantidad': 30}]
        self.ventas_totales = 0 

    def agregar_producto(self, nombre, precio, cantidad):
        encontrado = False

        for producto in self.inventario:
            if producto['nombre'] == nombre:  #esta el poroducto?
                producto['cantidad'] += cantidad  #sumamos la cantidad
                encontrado = True
                print(f'Nuevo stock:, {producto['cantidad']} de {nombre}')
                break   #para que corte la iteracion
    
        if not encontrado: 
            nuevo_producto = {'nombre': nombre, 'precio': precio, 'cantidad': cantidad}
            self.inventario.append(nuevo_producto)
            print(f"A침adido: {nombre} se ha guardado en el inventario.")
            


mi_tienda = TiendaOnline()
mi_tienda.agregar_producto ("pantalon", 4, 30)
mi_tienda.agregar_producto ('pantalon',4, 5)
mi_tienda.agregar_producto ('camisa', 4, 8)

    
```
