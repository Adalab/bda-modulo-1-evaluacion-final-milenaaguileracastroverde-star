```python
class TiendaOnline:
    def __init__(self):
        self.inventario = []  # inventario = [{'nombre': 'Camisa', 'precio': 20, 'cantidad': 40}, {'nombre': 'Pantalón', 'precio': 30, 'cantidad': 30}]
        self.ventas_totales = 0 

    def agregar_producto(self, nombre, precio, cantidad):
        encontrado = False

        for producto in self.inventario:
            if producto['nombre'] == nombre:  #esta el producto?
                producto['cantidad'] += cantidad  #sumamos la cantidad
                encontrado = True
                print(f'Nuevo stock:, {producto['cantidad']} de {nombre}') #imprime la suma de articulo
                break   #para que corte la iteracion si lo encuentra
    
        if not encontrado: 
            nuevo_producto = {'nombre': nombre, 'precio': precio, 'cantidad': cantidad}
            self.inventario.append(nuevo_producto)
            print(f"Añadido: {nombre} se ha guardado en el inventario.")

    def ver_inventario (self): #ver inventario
        for producto in self.inventario:
            print(f"Nombre: { producto['nombre'] }, Precio: { producto['precio'] }€, Cantidad: { producto['cantidad'] }") #Nombre: Camisa, Precio: $20, Cantidad: 50
            
    def buscar_producto(self, nombre):
        for producto in self.inventario:
            if producto['nombre'] == nombre: #compara nombres
                print(f"Nombre: { producto['nombre'] }, Precio: { producto['precio'] }€, Cantidad: { producto['cantidad'] }")
                break
    
    def actualizar_stock(self, nombre, cantidad): #Se repite la funcion
        encontrado = False
        for producto in self.inventario:
            if producto['nombre'] == nombre:  
                producto['cantidad']= cantidad  
                encontrado = True
                break

        if not encontrado:
            print('No encontrado')

 


            


mi_tienda = TiendaOnline()
mi_tienda.agregar_producto ('corbata', 9, 30)
mi_tienda.agregar_producto ('pantalon',7, 5)
mi_tienda.agregar_producto ('camisa', 4, 8)
mi_tienda.agregar_producto ('pantalon', 6, 3)


# mi_tienda.ver_inventario() 

# mi_tienda.buscar_producto('perro')

mi_tienda.actualizar_stock ('zapallo', 30)

    
```
